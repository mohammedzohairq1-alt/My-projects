## --> Start Data Cleaning

```python
df.drop(['Zipcode'],axis = 1,inplace = True)
```

## --> Let's Start Analyses

```python
ax = sns.countplot(x= 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
    ax.set_title('Gender Count', fontsize = 15)
    ax.set_xlabel('Gender %', fontsize = 10)
    ax.set_ylabel('Gender Count', fontsize = 10)
    ax.tick_params(axis = 'both',labelsize = 10)
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()
plt.pie(
    gender_counts,
    labels = gender_counts.index,
    autopct = '%1.2f%%',
    startangle = 90,
)
plt.title('Gender wise in pie', fontsize = 10)
plt.show()
```

```python
value_count = df.groupby('Gender',as_index = False)['Amount'].sum().sort_values(by= 'Amount',ascending = False)
ax = sns.barplot(x = 'Gender', y = 'Amount', data = value_count)
for bars in ax.containers:
    ax.bar_label(bars, fmt = '%.0f',fontsize= 15, color = 'Green')
plt.title('Gender Wise Amount',fontsize = 15)
plt.show()
```

```python
ax = sns.countplot(data = df, x= 'Age Group')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sort_values = df['Age Group'].value_counts().sort_values(ascending = True)
sorted_values = sort_values.index
ax = sns.countplot(data = df, x = 'Age Group', order = sorted_values)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sort_gender = df['Age Group'].value_counts().sort_values(ascending = True)
sorted_gender = sort_gender.index
ax = sns.countplot(data = df, x= 'Age Group', order= sorted_gender, hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
Total_amount = df.groupby(['Age Group'],as_index = False,)['Amount'].sum().sort_values(by= 'Amount', ascending= True)
sns.set(rc={'figure.figsize':[15,5]})
sns.lineplot(x= 'Age Group', y= 'Amount',data = Total_amount)
plt.show()
```

```python
state_orders = df.groupby(['State'],as_index =False)['Orders'].sum().sort_values(by= 'Orders', ascending= False).head(5)
sns.set(rc={'figure.figsize':[15,5]})
ax = sns.lineplot(x= 'State',y= 'Orders', data= state_orders)
for bars in ax.containers:
    ax.bar_label(bars)
    ax.set_xlabel('STATES',fontsize= 10)
    ax.set_ylabel('ORDERS',fontsize=10)
plt.title('Top 10 States',fontsize=25)
plt.show()
```

```python
Amount_wise = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(5)
sns.set(rc={'figure.figsize':[15,5]})
sns.barplot(data = Amount_wise, x= 'State', y= 'Amount')
plt.show()
```

```python
ax = sns.countplot(data = df,x= 'Product_Category')
sns.set(rc={'figure.figsize':[10,5]})
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('PRODUCT CATRGORY')
plt.show()
```

```python
Product_Category = df.groupby(['Product_Category'],as_index= False)['Amount'].sum().sort_values(by = 'Amount', ascending= False)
sns.set(rc={'figure.figsize':[15,5]})
sns.barplot(data = Product_Category, x= 'Product_Category',y= 'Amount')
plt.show()
```

```python
Product_gender = df.groupby(['Product_Category','Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(data = Product_gender, x= 'Product_Category',y= 'Amount', hue= 'Gender')
plt.show()
```

```python
ax = sns.countplot(data= df, x= 'Profession')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('BUYERS PROFESSIONS', fontsize= 15)
plt.show()
```

```python
profession_wise_Amount = df.groupby(['Profession'],as_index = False)['Amount'].sum().sort_values(by ='Amount', ascending = False).head(10)
sns.lineplot(x= 'Profession', y= 'Amount', data = profession_wise_Amount)
plt.title('Profession_Wise_Amount',fontsize = 15)
plt.show()
```
